```python
#change budget and gross to integer
movies_df["budget"] = movies_df["budget"].astype("int64")
movies_df["gross"]= movies_df["gross"].astype("int64")
```

```python
plt.figure(figsize =(9,6))

plt.scatter(x=movies_df["budget"], y=movies_df["gross"])
plt.title("Budget vs Gross Earning")
plt.xlabel("Gross Earning")
plt.ylabel("Budget for Film")
plt.show
```

```python
sns.regplot(x="budget", y="gross", data = movies_df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
correlation_matrix= movies_df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show
```

```python
num_df = movies_df

for col in num_df.columns:
    if(num_df[col].dtype == "object"):
        num_df[col]=num_df[col].astype("category")
        num_df[col]=num_df[col].cat.codes
```

```python
num_df.head()
```

```python
plt.figure(figsize =(12,6))

correlation_matrix= num_df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show
```

```python
sort_corr = num_df.corr().unstack()
sort_corr
```

```python
sort_pairs= sort_corr.sort_values()
sort_pairs
```

```python
high_corr = sort_pairs[(sort_pairs) >0.5]
high_corr
```
